# GENERADOR DE IMÁGENES: Figuras Geométricas Aleatorias

## Concepto
Este notebook implementa un generador de imágenes que crea composiciones artísticas 
con figuras geométricas aleatorias. Cada imagen es única y generada proceduralmente.

**Características:**
- Generación de imágenes RGB (color verdadero)
- Figuras: rectángulos, elipses, líneas y polígonos
- Colores aleatorios en formato RGB
- Posiciones y tamaños variables
- Múltiples formatos de salida (PNG, JPG)

**Formato de imagen:** PNG (Portable Network Graphics)
- **Ventajas:** Sin pérdida de calidad, transparencia, compresión sin artefactos
- **Resolución:** 800x600 píxeles (personalizable)
- **Profundidad de color:** 24 bits (16.7 millones de colores)

**Aplicaciones:**
- Arte generativo
- Fondos de pantalla únicos
- Material para pruebas de visión por computadora
- Inspiración para diseño gráfico
