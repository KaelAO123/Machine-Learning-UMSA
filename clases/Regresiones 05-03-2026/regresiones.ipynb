# Regresion Lineal

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
y = np.array([2, 4, 5, 4, 5])
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
X_fit = np.linspace(1,5,100).reshape(-1,1)
y_fit = model.predict(X_fit)
```

```python
plt.scatter(x,y,color="blue",label="datos")
plt.plot(X_fit,y_fit,color='red',label='Regresion Lineal')
plt.xlabel('X')
plt.ylabel("Y")
plt.title("Regresion Lineal Simple")
plt.legend()
```

# Regresion Logistica

```python
from sklearn.linear_model import LogisticRegression
```

```python
X = np.array([150,160,170,180,190]).reshape(-1,1)
y = np.array([0,0,1,1,1])
```

```python
log_model = LogisticRegression()
log_model.fit(X,y)
```

```python
X_fit = np.linspace(140,200,100).reshape(-1,1)
y_prob = log_model.predict_proba(X_fit)[:,1]
```

```python
plt.scatter(X,y,color="blue",label="datos")
plt.plot(X_fit,y_prob,color="purple",label="Probabilidad predicha")
plt.xlabel("Altura (cm)")
plt.ylabel("Probabilidad de practicar deporte")
plt.title("Regresion Logistica")
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
X = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([1,4,9,16,25])
degree = 2
poly_model = make_pipeline(PolynomialFeatures(degree),LinearRegression())
poly_model.fit(X,y)
X_fit = np.linspace(1,5,100).reshape(-1,1)
y_fit = poly_model.predict(X_fit)
plt.scatter(X,y,color="blue",label="datos")
plt.plot(X_fit,y_fit,color="green",label=f"Regresion Polinomial (grado {degree})")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Regresion Polinomial")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
np.random.seed(0)
X = np.linspace(0, 10, 100)
y = 0.5 * X**2 + X + 2 + np.random.randn(100) * 4
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
model = make_pipeline(PolynomialFeatures(2), Ridge(alpha=1.0))
model.fit(X_train.reshape(-1, 1), y_train)
```

```python
X_fit = np.linspace(0, 10, 100)
y_pred = model.predict(X_fit.reshape(-1, 1))
plt.scatter(X_train, y_train, color='blue', label='Datos de entrenamiento')
plt.scatter(X_test, y_test, color='red', label='Datos de prueba')
plt.plot(X_fit, y_pred, color='green', label='Modelo de Regresión Polinomial con Ridge')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresión Polinomial con Regularización Ridge')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
np.random.seed(0)
X = np.linspace(0, 10, 100)
y = 0.5 * X**2 + X + 2 + np.random.randn(100) * 4
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
ridge_model = make_pipeline(PolynomialFeatures(2), Ridge(alpha=1.0))
lasso_model = make_pipeline(PolynomialFeatures(2), Lasso(alpha=0.1))
elastic_model = make_pipeline(PolynomialFeatures(2), ElasticNet(alpha=0.1, l1_ratio=0.5))
ridge_model.fit(X_train.reshape(-1, 1), y_train)
lasso_model.fit(X_train.reshape(-1, 1), y_train)
elastic_model.fit(X_train.reshape(-1, 1), y_train)

X_fit = np.linspace(0, 10, 100)
y_ridge= ridge_model.predict(X_fit.reshape(-1, 1))
y_lasso = lasso_model.predict(X_fit.reshape(-1, 1))
y_elastic = elastic_model.predict(X_fit.reshape(-1, 1))
plt.scatter(X_train, y_train, color='blue', label='Datos de entrenamiento')
plt.scatter(X_test, y_test, color='red', label='Datos de prueba')
plt.plot(X_fit, y_ridge, color='green', label='Ridge')
plt.plot(X_fit, y_lasso, color='orange', label='Lasso')
plt.plot(X_fit, y_elastic, color='purple', label='ElasticNet')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Comparación: Ridge Lasso ElasticNet')
plt.legend()
plt.show()
```
