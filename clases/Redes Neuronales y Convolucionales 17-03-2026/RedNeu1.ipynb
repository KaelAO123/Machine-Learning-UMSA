```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Datos de entrada para XOR
X = np.array([[0,0],[0,1],[1,0],[1,1]])
# Salidas esperadas para XOR
y = np.array([[0],[1],[1],[0]])
```

```python
# Crear el modelo secuencial (red neuronal simple)
model = Sequential()
# Capa oculta con 16 neuronas y función ReLU
model.add(Dense(16, input_dim=2, activation='relu'))
# Capa de salida con 1 neurona y activación sigmoide
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compilar el modelo con optimizador Adam y función de pérdida binaria
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Entrenar la red
history = model.fit(X, y, epochs=1000, verbose=0)
```

```python

# Evaluar el resultado
loss, accuracy = model.evaluate(X, y, verbose=0)
print(f"Precisión final: {accuracy:.2f}")
```

```python
# Predecir con la red entrenada
predictions = model.predict(X).round()
print("Predicciones:")
print(predictions)
```

```python
# Graficar la pérdida durante el entrenamiento
plt.plot(history.history['loss'])
plt.title('Pérdida durante el entrenamiento')
plt.xlabel('Época')
plt.ylabel('Pérdida')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
```

```python
# Datos de entrada (XOR)
X = np.array([[0,0], [0,1], [1,0], [1,1]])
# Etiquetas (0 o 1)
y = np.array([0, 1, 1, 0])
# Convertir etiquetas a categorías (one-hot)
y_cat = to_categorical(y, num_classes=2)
```

```python
# Crear modelo secuencial con una capa oculta y capa de salida softmax
model = Sequential([
    Dense(4, input_dim=2, activation='relu'),
    Dense(2, activation='softmax')
])
```

```python
# Compilar el modelo con pérdida categórica y métrica de precisión
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Entrenar el modelo guardando historial
history = model.fit(X, y_cat, epochs=300, verbose=0)
```

```python
# Graficar evolución de pérdida y precisión
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='Pérdida')
plt.xlabel('Épocas')
plt.ylabel('Pérdida')
plt.title('Pérdida durante entrenamiento')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='Precisión')
plt.xlabel('Épocas')
plt.ylabel('Precisión')
plt.title('Precisión durante entrenamiento')
plt.legend()

plt.show()
```

```python
# Mostrar predicciones probabilísticas y clases predichas
predicciones = model.predict(X)
print("Predicciones (probabilidades):\n", predicciones)
print("Clases predichas:", np.argmax(predicciones, axis=1))
```
