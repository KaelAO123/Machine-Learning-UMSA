se comparan dos versiones de un elemento (version A, version B)
para determinar cual tiene una mayor tasa de conversion
la tasa de conversion se refiere a la proporcion de usuarios que realizan una accion
deseada (como hacer clic en un boton, o copletar una compra, etc)

version A, version original o de control, con una tasa de conversion del 5%
version B, variante, con una tasa de conversion del 6%

objetivo: analizar la direncia entres estas tasas de conversion si es significativa
si la version B realmente funciona mejor que la version A en terminos de
conversion. Se uso una prueba de proporciones (Z-test) para comparar tasas
de conversion de ambas versiones

```python
# simula una prueba A/B para comparar las tasas de conversion de dos versiones de un producto
# o servicio
# importamos las librerias
import numpy as np  #operaciones numericas
```

```python
# Simulación de datos
np.random.seed(0)  # Para reproducibilidad, una semilla para generar num aleatorios
```

```python
# Versión A (Control): Tasa de conversión del 5%
conversion_A = np.random.binomial(n=1, p=0.05, size=1000) #simula experimentos binomiales donde n=1
                                                          # indica un solo ensayo por observacion
                                                          # p prob de exito y size el tamaño
```

```python
# Versión B (Variante): Tasa de conversión del 6%
conversion_B = np.random.binomial(n=1, p=0.06, size=1000)
```

```python
conversion_B
```

```python
# Cálculo de las tasas de conversión reales
tasa_A = np.mean(conversion_A)  #calcula el promedio de cada version
tasa_B = np.mean(conversion_B)
```

```python
print(f"Tasa de conversión Versión A: {tasa_A:.4f}")
print(f"Tasa de conversión Versión B: {tasa_B:.4f}")
```

```python
# Prueba de hipótesis para comparar las tasas de conversión
# H0: p_A = p_B  las tasas de conversion son iguales
# H1: p_A ≠ p_B  las tasas de conversion son diferentes

# Utilizando una prueba de proporciones (Z-test)
# calculo de Z-score
from scipy.stats import norm
```

```python
n_A = len(conversion_A)
n_B = len(conversion_B)
```

```python
p_A = tasa_A
p_B = tasa_B
```

```python
# se calcula la proporcion combinada de ambos grupos
p_pooled = ((n_A * p_A) + (n_B * p_B)) / (n_A + n_B)

# calcula el error estandar de la diferencia entre las proporciones
std_error = np.sqrt(p_pooled * (1 - p_pooled) * (1/n_A + 1/n_B))

# calcula la puntuacion Z, indica ctos errores estandar se encuentra la diferencia
# entre las proporciones respecto a cero
z_score = (p_B - p_A) / std_error
```

```python
# Nivel de significancia (alpha)
# significa que hay un 5% de prob de rechazo la H0 cuando es verdadera
alpha = 0.05
```

```python
# Valor crítico para una prueba bilateral
z_crit = norm.ppf(1 - alpha/2)

# calcula el valor critico para una prueba bilateral, que es el
# valor de Z mas alla del cual se rechaza la H0
```

```python
print(f"Z-score: {z_score:.4f}")
print(f"Valor crítico (Z): {z_crit:.4f}")
```

```python
if abs(z_score) > z_crit:
    print("Rechazamos la hipótesis nula. Las tasas de conversión son significativamente diferentes.")
else:
    print("No rechazamos la hipótesis nula. Las tasas de conversión no son significativamente diferentes.")

# si el valor de Z_score > Z_crit, se rechaza la H0, lo que indica que las tasas de conversion
# son estadisticamente diferentes
```
