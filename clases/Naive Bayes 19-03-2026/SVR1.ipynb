```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
```

```python
X = np.array([[1], [2], [3], [4], [5], [6], [7], [8]])  # Años de experiencia
y = np.array([30, 35, 40, 50, 60, 70, 80, 90])          # Salario
```

```python
svr = SVR(kernel='linear', C=1.0, epsilon=0.1)
```

```python

svr.fit(X, y)
```

```python
y_pred = svr.predict(X)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print("Linear SVR R2:", r2_score(y, y_pred))
```

```python
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, y_pred, color='red', linewidth=2, label='Modelo SVR')
plt.title('Regresión por Máquinas de Vectores de Soporte')
plt.xlabel('Años de Experiencia')
plt.ylabel('Salario (miles)')
plt.legend()
plt.show()
```
