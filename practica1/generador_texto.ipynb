# GENERADOR DE TEXTO: "Generador de Nombres de Productos"
## Concepto
El generador de texto crea nombres creativos para productos combinando aleatoriamente adjetivos y sustantivos de listas predefinidas. Es útil para generar ideas de nombres de productos, marcas o campañas de marketing de manera rápida y automatizada.

**Formato de salida:**
- Texto en consola
- Archivo .txt (formato de texto plano UTF-8)

**Características:**
- Generación aleatoria de combinaciones
- Interfaz interactiva por consola
- Exportación a archivo de texto
- Validación de entrada del usuario

```python
import random  
from datetime import datetime 
```

## Librerías utilizadas

### 1. **random** (Librería estándar)
- **Propósito**: Generar números aleatorios y hacer selecciones al azar
- **Funciones principales**:
  - `random.choice(lista)`: Selecciona un elemento aleatorio de una lista
  - `random.randint(a, b)`: Genera un número entero aleatorio entre a y b

### 2. **datetime** (Librería estándar)
- **Propósito**: Manejo de fechas y horas
- **Uso en este programa**: Agregar timestamp a los archivos generados

```python
def generar_nombre_producto():
    """
    Genera un nombre aleatorio combinando adjetivo + sustantivo
    
    Returns:
        str: Nombre generado (ej. "Super Widget")
    """
    
    adjetivos = [
        "Super", "Mega", "Ultra", "Inteligente", "Eco", "Tech", 
        "Pro", "Premium", "Basico", "Deluxe", "Mini", "Maxi",
        "Hyper", "Power", "Quick", "Easy", "Fast", "Cool"
    ]
    
    sustantivos = [
        "Widget", "Gadget", "Herramienta", "Dispositivo", "App", "Sistema",
        "Solucion", "Caja", "Kit", "Pack", "Set", "Bundle",
        "Hub", "Pod", "Core", "Sync", "Link", "Connect"
    ]
    
    adjetivo = random.choice(adjetivos)
    sustantivo = random.choice(sustantivos)
    
    return f"{adjetivo} {sustantivo}"

def generar_multiples_productos(cantidad):
    """
    Genera múltiples nombres de productos
    
    Args:
        cantidad (int): Número de nombres a generar
    
    Returns:
        list: Lista con los nombres generados
    """

    return [generar_nombre_producto() for _ in range(cantidad)]

def guardar_en_archivo(productos, nombre_base="productos_generados"):
    """
    Guarda los productos en un archivo con timestamp
    
    Args:
        productos (list): Lista de productos a guardar
        nombre_base (str): Nombre base del archivo
    
    Returns:
        str: Nombre del archivo creado
    """

    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    nombre_archivo = f"{nombre_base}_{timestamp}.txt"
    
    with open(nombre_archivo, 'w', encoding='utf-8') as f:
        f.write("=" * 50 + "\n")
        f.write("PRODUCTOS GENERADOS\n")
        f.write(f"Fecha: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
        f.write("=" * 50 + "\n\n")
        
        for i, producto in enumerate(productos, 1):
            f.write(f"{i:3d}. {producto}\n")
        
        f.write(f"\nTotal: {len(productos)} productos")
    
    return nombre_archivo

def mostrar_productos_en_tabla(productos):
    """
    Muestra los productos en formato tabla para mejor visualización
    
    Args:
        productos (list): Lista de productos a mostrar
    """
    
    print("\n" + "=" * 60)
    print("PRODUCTOS GENERADOS".center(60))
    print("=" * 60)
    
    columnas = 3
    for i in range(0, len(productos), columnas):
        fila = productos[i:i+columnas]
        elementos_fila = []
        for j, prod in enumerate(fila):
            num = i + j + 1
            elementos_fila.append(f"{num:2d}. {prod:15s}")
        print("   ".join(elementos_fila))
    
    print("=" * 60)
    print(f"Total: {len(productos)} productos")
```

```python
def main():
    """Función principal del programa"""
    
    print("=" * 60)
    print("GENERADOR DE NOMBRES DE PRODUCTOS".center(60))
    print("=" * 60)
    
    while True:
        print("\nOPCIONES:")
        print("   1. Generar nombres")
        print("   2. Salir")
        
        opcion = input("\nSelecciona una opción (1-2): ").strip()
        
        if opcion == "1":
            try:
                cantidad = int(input("   ¿Cuántos nombres generar? "))
                
                if cantidad <= 0:
                    print("Debe ser un número positivo")
                    continue
                
                print(f"\nGenerando {cantidad} nombres...")
                
                productos = generar_multiples_productos(cantidad)
                
                mostrar_productos_en_tabla(productos)
                
                guardar = input("\nGuardar en archivo? (s/n): ").lower()
                if guardar == 's':
                    archivo = guardar_en_archivo(productos)
                    print(f"Guardado en: {archivo}")
                
            except ValueError:
                print("Error: Ingresa un número válido")
            except Exception as e:
                print(f"Error inesperado: {e}")
        
        elif opcion == "2":
            print("\nHasta luego!")
            break
        
        else:
            print("Opción no válida")

if __name__ == "__main__":
    main()
```
